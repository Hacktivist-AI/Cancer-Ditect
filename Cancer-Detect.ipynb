```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('cancer.csv')
```

```python
data=data.drop(['Unnamed: 32','id'],axis=1)
data.diagnosis=[1 if each == "M" else 0 for each in data.diagnosis ]
```

```python
data.diagnosis
```

```python
y=data.diagnosis
x0=data.drop(['diagnosis'],axis= 1)
```

```python
#d=np.array([0,10,20,200,40,50,100])
```

```python
x=(x0 - np.min(x0)) / (np.max(x0) - np.min(x0))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state= 42)
```

```python
from sklearn import linear_model
robot=linear_model.LogisticRegression()

robot.max_iter=1000000
robot.fit(x_train,y_train)
```

```python
print("x_train:",x_train.shape)
print("x_test:",x_test.shape)
print("y_train:",y_train.shape)
print("y_test:",y_test.shape)
```

```python
t=robot.predict(x_test)
v= y_test ==t
v= v.reset_index()
v= v.diagnosis
n= 0  
for i in v:
    if i ==False : n=n+1
print("Tedad_khata: ",n,'/',len(v)) 
print("Darsad_khata: ",(n * 100)/len(v))
print("Deggat_Dorosti: ",100-(n * 100)/len(v))   
```
